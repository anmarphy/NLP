```python
soup = BeautifulSoup(response.content, 'html.parser')

title = soup.find(id="firstHeading")
print(title.string)
```

```python
tot_parag=len(soup.find_all('p'))
tot_parag
```

```python
l=[]
for i in range(0, tot_parag):
    text=soup.find_all('p')[i].get_text()
    l.append(text)
l
```

```python
text= ''
for i in range(0, len(l)):
    text=text+' '+ l[i]
```

```python
text
```

```python
def process_text(text):
    stopwords_english = stopwords.words('english')
    # remove stock market tickers like $GE
    text = re.sub(r'\$\w*', '', text)
    
    # remove hyperlinks
    text = re.sub(r'https?:\/\/.*[\r\n]*', '', text)
    
    # only removing the hash # sign from the word
    text = re.sub(r'#', '', text)
    
    #Removing new line
    text=re.sub(r'\n', '', text )
    
    #Removing references
    text=re.sub(r'\[[0-9+]+]', '', text )
    
    #Replacing \'s by s
    text=re.sub(r"\'s" , "s", text )
    
    #Removing text in parenthesis
    text=re.sub("[\(\[].*?[\)\]]", '', text)
    
    
    return text
```

```python
text_clean=process_text(text)
text_clean[0:1000]
```

```python
custom_mask = np.array(Image.open("meditate.jpg"))
wc = WordCloud(background_color="white", mask=custom_mask)
wc.generate(text_clean)
plt.figure(figsize=(10,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

## Name Entity Recognition

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
from spacy import displacy
doc = nlp(text_clean)
displacy.serve(doc, style="ent")
```
